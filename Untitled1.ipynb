```python
from google.colab import files
import os
import zipfile

uploaded = files.upload()
zip_path = list(uploaded.keys())[0]

extract_folder = "faces_dataset"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print(f"✅ Dataset extracted to folder: {extract_folder}")
print("Users:", os.listdir(extract_folder))

import face_recognition
import pickle
import datetime

def train_face_encodings(dataset_path="faces_dataset", model_save_path="encodings.pkl"):
    known_encodings = []
    known_names = []

    for user_name in os.listdir(dataset_path):
        user_folder = os.path.join(dataset_path, user_name)
        if not os.path.isdir(user_folder):
            continue
        for file in os.listdir(user_folder):
            img_path = os.path.join(user_folder, file)
            image = face_recognition.load_image_file(img_path)
            encodings = face_recognition.face_encodings(image)
            if len(encodings) > 0:
                known_encodings.append(encodings[0])
                known_names.append(user_name)

    with open(model_save_path, "wb") as f:
        pickle.dump({"encodings": known_encodings, "names": known_names}, f)

    print(f"[✅] Training complete! Encodings saved to {model_save_path}")

def verify_user(username, image_path, model_path="encodings.pkl", tolerance=0.5):
    with open(model_path, "rb") as f:
        data = pickle.load(f)
    unknown_image = face_recognition.load_image_file(image_path)
    unknown_encodings = face_recognition.face_encodings(unknown_image)

    if len(unknown_encodings) == 0:
        return "No Face Detected", 0.0

    unknown_encoding = unknown_encodings[0]
    matches = face_recognition.compare_faces(data["encodings"], unknown_encoding, tolerance=tolerance)
    face_distances = face_recognition.face_distance(data["encodings"], unknown_encoding)

    if True in matches:
        best_match_index = face_distances.argmin()
        matched_name = data["names"][best_match_index]
        confidence = 1 - face_distances[best_match_index]

        if matched_name == username:
            return "Access Granted", float(confidence)
        else:
            return "Access Denied", float(confidence)
    else:
        return "Access Denied", 0.0


train_face_encodings(dataset_path=extract_folder, model_save_path="encodings.pkl")


print("\n=== Verifying all images in dataset ===")
for user_name in os.listdir(extract_folder):
    user_folder = os.path.join(extract_folder, user_name)
    if not os.path.isdir(user_folder):
        continue
    for file in os.listdir(user_folder):
        img_path = os.path.join(user_folder, file)
        status, confidence = verify_user(user_name, img_path)
        print(f"User: {user_name} | Image: {file} | Result: {status} | Confidence: {confidence:.2f}")


files.download("encodings.pkl")
```
